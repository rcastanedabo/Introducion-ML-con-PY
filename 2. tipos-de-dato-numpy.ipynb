### Cambiando tipo de datos

```python
arr = np.array([0,1,2,3,4])
arr = arr.astype(np.bool_)
arr
```

```python
arr = np.array([0,1,2,3,4])
arr = arr.astype(np.string_)
arr
```

aquí, definí el tipo de dato, directamente desde la librería de numpy, le coloca esa b delante porque al ser un número me indica que el tipo es string

```python
arr = np.array(['0','1','2','3','4'])
arr = arr.astype(np.int8)
arr
```

Aquí, estoy haciendo al revés, ya que pongo los número como enteros y luego le digo que le tipo de datos es entero y funciona
