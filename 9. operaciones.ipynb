```python
matriz = arr.reshape(2,5)
matriz2 = matriz.copy()
```

```python
matriz
```

```python
matriz + matriz2
```

**multiplicación de dos matrices**

debemos de asegurarnos que se puedan multiplicar

```python
matriz2.T
```

en este caso a la matriz2 le aplicamos la traspuesta para que puedan multiplicarse

```python
np.matmul(matriz,matriz2.T)
```

una forma directa, sería aplicar la instrucción matmul de numpy pero no descuidando que la matriz2 se le ponga la traspuesta para que pueda multiplicarse.

```python
matriz @ matriz2.T
```

Otra forma igual, sería con el arroba @ pero también añandiendo la traspuesta a matriz2 para que pueda multiplicarse
