### Concatenate

```python
a = np.array([[1,2],[3,4]])
b = np.array([5,6])
```

creamos dos arrays: a de 2 x 2, b 

```python
np.concatenate((a,b),axis=0)
```

el error que me da es debido a que las dimensiones son distintas y por ello no podré concatenar

```python
a.ndim
```

con ndim veo la dimensión que tienen el array a, me sale 2

```python
b.ndim
```

por el contrario, el array b me sale 1. 

Es por ello que no puedo concatenar, ambos arrays deben de tener la misma dimensión para ello usamos expand. 

```python
b = np.expand_dims(b,axis=0)
```

expando la dimensión del array b, por el eje 0 y lo volvemos a guardar en b para que se actualice

```python
np.concatenate((a,b),axis=0)
```

Vuelvo a utilizar la instrucción concatenar (por el eje 0) y comprobamos que concateno ambos arrays

```python
b
```

```python
np.concatenate((a,b),axis=1)
```

que pasa si, por el contrario, quiero concatenar por eje 1, me da error 

```python
np.concatenate((a,b.T),axis=1)
```
