### Importar librerías

Dos funciones importantes de los arreglos: 
- Shape: me indica la forma que tiene un arreglo, qué estructura estoy trabajando, cómo se está comportando, cómo puedo ingresar a él, como manipularlo.


```python
import numpy as np 
```

```python
arr = np.random.randint(1,10,(3,2))
```

generamos un array aleatorios de 1 a 10, con una estructura de 3 x 2. Lo guardamos dentro de un array

### Shape

```python
arr.shape
```

aquí me devuelve, cuántas filas y cuántas columnas tengo

```python
arr
```

### Reshape

```python
arr.reshape(1,6)
```

reshape, cambia la forma como estuvo antes. en el ejemplo arr era de 3x2 con el reshape pasa a 1x6

```python
arr.reshape(2,3)
```

```python
np.reshape(arr,(1,6))
```

```python
arr
```

```python
np.reshape(arr,(2,3),'C')
```

le voy a hacer un reshape de una estructura de 2 x 3 en el lenguaje C, lo ordena por filas

```python
np.reshape(arr,(2,3),'F')
```

Aquí lo ordena distinto, aquí con F, lo hace como fortran, lo ordena por columna

```python
np.reshape(arr,(2,3),'A')
```
